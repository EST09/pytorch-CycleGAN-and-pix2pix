```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
#read in a 256x256 image and convert to 512x512 and save it

base_folder = "/users/rittscher/qdv200/pytorch-CycleGAN-and-pix2pix/results/ihc2he_pretrained/test_latest/images/"
results_folder = "/well/rittscher/users/qdv200/MPN/GAN_tiles/extra_data/bETGAN_6CALR_fake_HE_for_ssd/"

for file in os.listdir(base_folder):
    if "fake" in file:
        #print(file)
        img = cv2.imread(base_folder + file)
        new_file = file.replace("_fake", "")
        #img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (512, 512))
        cv2.imwrite(results_folder + new_file, img)
```
